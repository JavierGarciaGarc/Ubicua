# Linear Regression

Metodologa de aprendizaje para estimar el valor real de salida en base a un valor de entrada nuevo.



*   Introduccin con un modelo sencillo
*   Funcin de coste usada: MSE
*   Mtodo de Aprendizaje: *Gradient Descent*



---

Un modelo lineal sencillo lo podemos expresar como:

$$ h_{\theta}(x)= \theta_0 + \theta_1 x $$

donde,

* $h_{\theta}(x)$ representa nuestro modelo.
* $\hat{y}$  es el valor estimado y lo calculamos usando $h_{\theta}(x)$.
* $x$ representa el valor de la *feature*.
* $\theta$ son parmetros y son los valores que queremos ajustar. Al valor $\theta_0$ se le llama *bias*.

Tanto los parmetros $\theta$ como la feature $x$ podemos extenderlo a una cantidad de $n$ features. En este caso los podndramos de forma vectorial:

\begin{equation*}
   \hat{y} =  h_{\boldsymbol{\theta}}(\mathbf{x}) =\boldsymbol{\theta}\cdot \mathbf{x}
\end{equation*}

## Generacin de datos como ejemplo

Vamos a explorar la idea del modelo de regresin lineal, comenzando por generar datos con ruido aadido.

```
npr.seed(0)

x = np.linspace(-5, 5, 50)
tt1_real = 2
tt0_real = -5

def noise(t):
  return npr.normal(0, 1, size=t)

y = tt1_real * x + tt0_real + noise(len(x))

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y');
# en caso de necesitar guradar la figura, usar savefig()
# plt.savefig('miLR.pdf', format='pdf', dpi=1000, bbox_inches='tight')
```

Nota: Los valores de $x$ estn equiespaciados pero podra ser cualquier valor.

Ahora vamos a dar valores a una hiptesis cualquiera y representamos respecto a nuestros valores:

```
# establecemos una hiptesis cualquiera
tt1 = 1
tt0 = -8
y_est = tt1 * x + tt0

# plot del traning set
plt.scatter(x, y, label='datos')
#plot de la hiptesis
plt.plot(x, y_est, color='red', label='hiptesis')
plt.xlabel('x')
plt.ylabel('y')
plt.legend();
```

Lo que debe realizar nuestro modelo de aprendizaje es minimizar el *Mean Square Error*, MSE, y la podemos definir como:

$$\text{MSE} = \frac{1}{2m} \sum_{i=1}^{m}(h_\theta(x^{(i)})-y^{(i)})^2$$

Esta funcin ser nuestra funcin de coste y la representamos por $J(\theta_0, \theta_1)$, o, smplemente, $J(\boldsymbol{\theta})$ con

\begin{align}
    \boldsymbol{\theta} &= \begin{bmatrix}
           \theta_{0} \\
           \theta_{1}
         \end{bmatrix}
\end{align}

es decir, $J(\boldsymbol{\theta}) = \text{MSE}$.
(La terminologa de Machine Learning tambin podemos encontrar que a MSE le llama como *loss* o prdida).

El valor de $(h_\theta(x^{(i)})-y^{(i)})$ es el **residuo** de cada muestra $i$ del *training set*.

En este ejemplo estamos usando slo una *feature* y tenemos que ajustar slo dos valores de $\theta$. En general, la funcin de coste se puede extender a $n$ features y $n+1$ parmetros (incluyendo $\theta_0$).

Vamos a implementar una funcin que calcule MSE.

```
# Ejercicio:
# La funcin debe tomar como parmetros de entrada
# y (y_real) y la y_est (y_estimada)
# Debe devolver un escalar con el MSE.

def mse(y_real: np.ndarray, y_pred: np.ndarray) -> np.ndarray:
  # Note q esta definicin de funcin es nueva a partir de Python 3.5
  # typing: comprobacin de tipos de la fun.
  # REALIZAR funcin:

  # una forma ms compacta:
  return (np.mean(np.power(y_pred - y_real, 2)) / 2 )

mse(y, y_est)
```

## Optimizacin *manual* de MSE

Ahora aadimos unos sliders a la figura de arriba para tratar de encontrar el mejor ajuste a los datos del trining set:

```
def opt_plot(tt1, tt0):
    y_est = tt0 + tt1 * x
    plt.scatter(x, y, alpha=0.5, label='datos')
    plt.plot(x, y_est, color='red', label='hiptesis')
    plt.xlabel(r'$x$')
    plt.ylabel(r'$y$')
    plt.legend()
    plt.title(f'MSE: {mse(y, y_est):.02f}')
    plt.show()

# Define sliders
tt0_slider = FloatSlider(value=0, min=-10, max=10, step=0.1, description='tt0')
tt1_slider = FloatSlider(value=0, min=-10, max=10, step=0.1, description='tt1')

# Create interactive plot
interact(opt_plot, tt0=tt0_slider, tt1=tt1_slider)
```

Por tanto, nuestro objetivo es encontrar los parmetros $\boldsymbol{\theta}$ que minimicen el valor de nuestra funcin de coste, y lo podemos expresar como:

$$
  \underset{\boldsymbol{\theta}}{\operatorname{argmin}} J(\boldsymbol{\theta})
$$

Podemos sustituir la funcin de coste por su expresin general para $n$ features. En este caso la hiptesis $h_{\boldsymbol{\theta}}(x)$ podemos espresarla en forma de vectorial como $h({\bf x}) = \boldsymbol{\theta}^{\sf T}{\bf x}$, donde

\begin{align}
    \boldsymbol{\theta} &= \begin{bmatrix}
           \theta_{0} \\
           \theta_{1} \\
           \vdots \\
           \theta_{n}
         \end{bmatrix} , \,\in {\mathbb{R}}^{n + 1} \,,&  
    {\bf x} = \begin{bmatrix}
           x_{0} \\
           x_{1} \\
           x_{2} \\
           \vdots \\
           x_{n}
         \end{bmatrix}  , \,\in {\mathbb{R}}^{n+1}
\end{align}

Por tanto, la funcin de coste a optimizar queda de la siguiente forma:

$$
  J(\boldsymbol{\theta}) = \frac{1}{2m} \sum_{i=1}^{m} \left(\boldsymbol{\theta}^{\sf T}{\bf x}^{(i)} - y^{(i)} \right)^2
$$

Para resolver este problema de optimizacin de la funcin de coste se utiliza el mtodo de **Gradient Descent**.

## Algoritmo de Gradient Descent

La idea tras Gradient Descent es la de ir iterando para ir variando los valores de los parmetros $\boldsymbol{\theta}$ hasta encontrar un que haga que la derivada de la funcin de coste sea igual a cero, y por tanto ser un valor mnimo. El valor del vector $\boldsymbol{\theta}$ en ese mnimo nos resuelve el problema de optimizacin.

El algoritmo puede converger a un mnimo global o tambin a uno local.

El algoritmo es el siguiente:

Repeat:

{
\begin{align}
\theta_{j} &\leftarrow \theta_{j} - \alpha\frac{\partial}{\partial \theta_j}J(\boldsymbol{\theta}) \\ & \forall j=0,\ldots,n
\end{align}
  }

donde $\alpha$ es un parmetro que regula la *velocidad de aprendizaje* del algoritmo. Un valor de $\alpha \approx 0$ har que converja muy lentamente. Un valor relativamente grande de $\alpha$ puede provocar *overshooting* he incluso hacer que el algoritmo diverja.

El valor del gradiente de $J(\boldsymbol{\theta})$ queda:

$$
\frac{\partial}{\partial \theta_j}J(\boldsymbol{\theta})= \frac{1}{m}\sum_{i=1}^{m} \left(\boldsymbol{\theta}^{\sf T}{\bf x}^{(i)} - y^{(i)} \right)x_j^{(i)}
$$

La idea es ir dando pequeos pasos de la funcin **siempre** en la direccin donde $\theta$ hace que la funcin de coste tenga un mnimo:

- Si la derivada $\partial/\partial \theta_j > 0$ el valor de $\theta_j$ decrecer, y,
- Si la derivada $\partial/\partial \theta_j < 0$ el valor de $\theta_j$ aumentar.

conforme nos aproximemos al valor mnimo, la derivada parcial ser menor y por tanto el avance ser ms lento. En el valor mnimo la derivada es = 0 y el valor de $\theta$ ya no cambia de una iteracin a otra.
La condicin de convergencia la podemos establecer cuando la variacin de $J(\boldsymbol{\theta})$ en la iteracin es $\epsilon \leq 10^{-3}$, denominado *tolerancia*.

### Ejercicio:

Como ejercicio podemos comprobar su funcionamiento aplicndolo al caso sencillo de una funcin con un mnimo. Por ejemplo, la siguiente funcin:

$$
f(w) = -5 + 3w + w^2
$$

El mnimo de funcin est en $w= -3/2$ y podemos calcularlo usando **Gradient Descent**. (Ponemos $w$ para no confundir con $x$ que es el smbolo usado para las *features*).

```
# funcin
def f(w: np.ndarray):
  """
  ejemplo de una funcin hiptesis h(w)
  """
  return (-5 + 3*w + w**2)

# funcin derivada de f con respecto a w.
def df(w):
  """
  Derivada de h(x) con respecto a x
  """
  return (2*w + 3)

#--------------------------------------
```

```
# Para aplicar el algoritmo escogemos un instante inicial
w = 3.14

alpha = 0.005 # learning rate
# Ahora ajustamos 1000 veces el valor de w, tomando pequeos pasos
# en la direccin del gradiente
lastts = []
for i in range(500):
    lastts.append(w)
    w = w - alpha * df(w) # \alpha = 0.01.

print(w)
print(f(w))
```

Una manera de saber si el algoritmo est funcionando correctamente es representar el valor de la funcin que queremos encontrar el mnimo respecto a las iteraciones.

```
plt.plot([f(t) for t in lastts], label='f(w)')
plt.xlabel('iteraciones')
plt.ylabel('f(w)')
plt.legend();
```

El valor mnimo de $f$ es $f(w=-3/2)=-7.25$.

De igual modo,  operaramos para encontrar el valor de $\boldsymbol{\theta}$ que hace mnima nuestra funcin de coste $J(\boldsymbol{\theta})$.

## Vectorizacin de Gradient Descent

El ajuste de los valores de los parmetros $\theta_j$ requiere que en cada iteracin calculemos la derivada parcial de la funcin de coste para cada uno de los parmetros.
En lugar de eso, podemos hacer el ajuste de todos los $\theta_j$ de una vez si usamos matrices.

Usamos un vector gradiente, $\nabla_{\boldsymbol{\theta}}J({\boldsymbol{\theta}})$ que contenga todas las derivadas parciales,


\begin{align}
    \nabla_{\boldsymbol{\theta}}J({\boldsymbol{\theta}}) &= \begin{bmatrix}
           \frac{\partial}{\partial \theta_0}J(\boldsymbol{\theta}) \\
           \frac{\partial}{\partial \theta_1}J(\boldsymbol{\theta}) \\
           \vdots \\
           \frac{\partial}{\partial \theta_n}J(\boldsymbol{\theta})
         \end{bmatrix} = \frac{1}{m}{\bf X^{\sf T}}\left( {\bf X}{\boldsymbol{\theta}} - {\bf y} \right)
\end{align}

Loa cmputos sobre la matriz ${\bf X}$ que representa el training set, hace que este algoritmo no escale bien en *training sets* grandes (en ese caso mejor usar la ecuacin normal que veremos despus). Sin embargo, s escala bien con el nmero de *features*.

El algoritmo queda de la siguiente forma:

Repeat:

$$
\boldsymbol{\theta} \leftarrow \boldsymbol{\theta} - \alpha \nabla_{\boldsymbol{\theta}}J({\boldsymbol{\theta}})
$$

<br>

Como el algoritmo requiere calculos con el training set completo en cada iteracin, a este algoritmo se le conoce como *Batch Gradient Descent*.
Tiene como desventaja que puede ser lento con training sets grandes. Sin embargo, **escala bien** con el nmero de *features*.


### Ejercicio:

Implementar Gradient Descent de forma vectorial/matricial.

```
# Con la finalidad de comparar vamos a usar
# los datos generados al principio del notebook
npr.seed(0)
m = 50 # nmero de muestras en el training set
x = np.linspace(-5, 5, m)
tt1_real = 2
tt0_real = -5

def noise(t):
  return npr.normal(0, 1, size=t)

y = tt1_real * x + tt0_real + noise(m)
#-----------------------------
# GRADIENT DESCENT
#-----------------------------

X = x[:,np.newaxis]
y = y[:, np.newaxis]
# debemos aadir el feature x_0=1 correspondiente al parmetro bias tt_0
# en cada instancia de X
X_b = np.c_[np.ones((len(x),1)), X]

# Gradient Descent:
# Nota: a la hora de aplicar el mtodo observe el formato de las matrices
# X debe ser (m x n+1)
# y debe ser (m x 1)
# \theta es  (n+1 x 1)
# en nuestro ejemplo n = ?

alpha = 0.1 # learning rate
max_iterations = 500

# inicializamos vector theta con un valor aleatorio
theta = -2 * npr.randn(2,1)
f_cost = []
for iteration in range(max_iterations):
  gradients = 1/m * X_b.T.dot(X_b.dot(theta) - y)
  theta = theta - (alpha * gradients)
  f_cost.append( mse( y, X_b.dot(theta)) )

print('Los valores ptimos de \u03B8 son: ', theta)
```

Observamos que los valores los los parmetros $\theta$ son parecidos a los que usamos inicialmente para generar los datos reales. El ruido aleatorio que aadimos hace variar los parmetros que consiguen minimizar la funcin de coste.

```
plt.plot(f_cost, label=r'$J(\theta)$')
plt.xlabel('iteraciones')
plt.ylabel(r'$J(\theta)$')
plt.legend();
```

En esta figura vemos que la funcin de coste decrece w.r.t iteraciones y que converge al valor mnimo. Una forma diferente, por ejemplo, si el valor de la funcin crece, indicara que debemos probar con una velocidad de aprendizaje, $\alpha$, menor.

## Mtodo Analtico: *Ecuacin Normal*

Adems de gradient descent existe una **forma directa** de calcular los parmetros $\boldsymbol{\theta}$ que minimizan la funcin de coste $J(\boldsymbol{\theta})$.

El resultado se llama **ecuacin normal** y es la siguiente expresin:

$$
\hat{\boldsymbol{\theta}} = \left( {\bf X}^{\sf T} {\bf X} \right)^{-1} {\bf X}^{\sf T} {\bf y}
$$


Donde:

- $\hat{\boldsymbol{\theta}}$ es el valor de $\boldsymbol{\theta}$ que minimiza la funcin de coste.
- ${\bf y}$ es el vector *target*, y contiene $y^{(1)}\ldots y^{(m)}$
- ${\bf X}$ representa el training set y tiene dimensin $m\; \textsf{x}\; (n+1) $. Es decir, aado una columna al principio de la matriz ${\bf X}$ con todo 1's



La principal ventaja de esta forma de calcular los parmetros es que

- No hay que iterar.
- No hay que determinar un valor correcto de $\alpha$.

Como desventaja

- No escala bien con $n$ (features) y puede resultar lento.
- Hay que calcular $\left( {\bf X}^{\sf T} {\bf X} \right)^{-1}$

<br>

**Nota**: No lo vamos a demostrar, pero la Ecuacin Normal se obtiene de resolver $\frac{\text{d} J(\boldsymbol{\theta})}{\text{d} \boldsymbol{\theta}}=0$ .

### Aplicacin


Calculemos los parmetros $\boldsymbol{\theta}$ de nuestro ejemplo usando la Ecuacin Normal:

```
# La ecuacin es una lnea de cdigo
# pero podemos usar una funcin con parmetros de entrada
# nuestro training set y el vector de target 'y'

def normal_equation(trainset, target):
  """
  training set: matriz de (m x (n+1))
  target:  vector (m x 1)
  """
  return np.linalg.inv(trainset.T.dot(trainset)).dot(trainset.T).dot(target)
```

```
theta_best = normal_equation(X_b, y)
print('\u03B8 = ', theta_best)
```

El resultado es el mismo que el obtenido con Gradient Descent.

## Prediccin de valores nuevos

Para obtener una prediccin con valores nuevos de las features debemos resolver la hiptesis para ese ejemplo, es decir, calcular:

$$
h_{\boldsymbol{\theta}}({\bf X}) = {\bf X}{\boldsymbol{\theta}}

## Polynomial Regression

Hasta ahora hemos resuelto problemas donde nuestros datos se comportan de forma lineal. Qu sucede si nuestros datos se comportan de una form no lineal?.

En este caso ms complejo tambin podemos aplicar un estimador lineal usando el siguiente *truco*:

* Podemos extender el training set, ${\bf X}$, con nuevas *features* que sean potencias o combinaciones de las features existentes.

A esta tcnica se le denomina **Polynomial Regression**.

Es posible extender el training set usando `sklearn` como vemos a continuacin.

### Ejemplo:

A modo de ejemplo, podemos geneerar un training set con datos no lineales basados en un polinomio de segundo grado (ecuacin cuadrtica), al que aadiremos algo de ruido.

```
npr.seed(0)
m = 100
X = 6 * npr.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + npr.randn(m ,1)
```

```
plt.scatter(X, y)
plt.xlabel('$x_{1}$', fontsize=14)
plt.ylabel('$y$', fontsize=14);
```

```
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```
X.shape
```

```
X_poly.shape
# En X_poly tenemos 2 columnas [X,X^2] -> degree=2
```

```
X_poly[0]
```

```
0.29288**2
```

Ahora podemos ajustar un modelo `LinearRegression` a este training set ampliado.

```
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
```

```
lin_reg.intercept_ , lin_reg.coef_
```

```
# generamos nuevos datos usando el objeto poly_features
# que hemos generado con nuestros datos
X_new = np.linspace(-3, 3, 100).reshape(100, 1)
# Para generar los datos de test usamos .tranform;
#.fit_transform es slo para el training set
X_new_poly = poly_features.transform(X_new)
y_new = lin_reg.predict(X_new_poly)
plt.scatter(X, y)
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predicciones")
plt.xlabel("$x_1$", fontsize=16)
plt.ylabel("$y$", rotation=0, fontsize=16)
plt.legend(loc="upper left", fontsize=14);
plt.axis([-3, 3, 0, 10])
#plt.show()
```

## Overfitting y Underfitting

Uno de los problemas que podemos encontrar es seleccionar el grado del polinmio que queremos usar como hiptesis para nuestros datos.

Si seleccionamos un grado que resulte *bajo* algunos datos nuevos (usamos para validar el modelo) ajustarn bien con el training set original y otros sin embargo, quedarn lejos y tendrn bastante error. En este caso decimos que hay **underfitting**.

Por otro lado, si el grado del polynomio de nuestra hiptesis es alto, ajustar muy bien los datos del training set pero no as los datos nuevos que usemos para validar. El polinmio hiptesis oscilar mucho entre los puntos y se alejar de los datos del training set, y, se producirn errores altos. En este caso decimos que hay **overfitting**.

Para ilustrar estos problemas ajustaremos los datos con un 3 tipos de polinmio de grado diferente (por ejemplo, 1, 2 y 100).

Adems, vamos a **normalizar** son la clase `StandardScaler` con el objetivo de que el algoritmo de optimizacin converja antes.

```
# "StandardScaler" es necesario para grado alto (=100).
# Equivale a "normalizar":
# vamos a tener features con valores mucho ms grandes que otros
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

for style, width, degree in (("g-", 1, 100), ("b--", 2, 2), ("r-+", 2, 1)):
    polybig_features = PolynomialFeatures(degree=degree, include_bias=False)
    std_scaler = StandardScaler()
    lin_reg = LinearRegression()
    polynomial_regression = Pipeline([
            ("poly_features", polybig_features),
            ("std_scaler", std_scaler),
            ("lin_reg", lin_reg),
        ])
    polynomial_regression.fit(X, y)
    y_newbig = polynomial_regression.predict(X_new)
    plt.plot(X_new, y_newbig, style, label=str(degree), linewidth=width)

plt.scatter(X, y, s=10)
plt.legend(loc="upper center")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([-3, 3, 0, 10]);
```

### Regularizacin

Tanto en problemas supervisados de regresin como en problemas de clasifficacin, si la cantidad de features es elevada se producir **overfitting**. La forma de solucionar este problema es realizando **Regularizacin**.

La idea tras la regularizacin no es reducir las features del training set, ya que, pueden ser relativamente importantes incluirlas en el modelo. Lo que se hace es **modular** su importancia haciendo que los parmetros $\theta_j$ correspondiente a la feature $j$ sea muy bajo.
